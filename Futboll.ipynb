```python
info_df = pd.DataFrame()
info_df["Id"] = df["Id"]
info_df["Name"] = df["Player"]
info_df["Nation"] = df["Nation"]
info_df["Team"] = df["Squad"]
info_df["League"] = df["Comp"]
info_df["Born"] = df["Born"]
info_df["Mac"] = df["MP"]
```

```python
info_df.head()
```

```python
df.drop(["Player","Nation","Squad","Comp","Born","MP"],inplace = True, axis = 1)
```

```python
for i in df.columns:
    if i == "Pos":
        continue
    else:
        df[i] = pd.to_numeric(df[i], errors='coerce')
```

```python
df.head()
```

```python
df.fillna(0.0, inplace = True)
```

```python
df.isnull().sum()
```

```python
for i in df:
    if i == "Pos":
        continue
    else:
        df[i] = df[i].astype(float)
```

```python
for i in df:
    if (i == "Pos") | (i == "Id"):
        continue
    else:
        df[f"{i}_s"] = MinMaxScaler(feature_range = (0,1)).fit(df[[i]]).transform(df[[i]])
```

```python
def_df = pd.DataFrame()
def_df["Id"] = df["Id"]
def_df["Pos"] = df["Pos"]
#Bloklar
def_df["Blok"]= df["Blocks_s"]*30/100 + df["BlkSh_s"]*20/100 + df["BlkShSv_s"]*25/100 + df["BlkPass_s"]*25/100  
#Müdahale
def_df["Baskı"]= df["Tkl_s"]*10/100 + df["TklWon_s"]*10/100 + df["TklDef3rd_s"]*14/100 +  \
df["TklDriAtt_s"]*12/100 + df["TklDri%_s"]*12/100 + df["Press_s"]*10/100 + \
df["PresDef3rd_s"]*14/100 + df["PresMid3rd_s"]*6/100 + df["TklW_s"]*12/100

def_df["Pas"] = df["PasTotPrgDist_s"]*8/100 + df["PasLonCmp_s"]*10/100 + df["PasAss_s"]*8/100 + df["Pas3rd_s"]*8/100 + \
df["PPA_s"]*10/100 + df["CrsPA_s"]*10/100 + df["PasAtt_s"]*10/100 + df["PasLive_s"]*10/100 + \
df["PasPress_s"]*10/100 + df["Sw_s"]*8/100 + df["Rec_s"]*8/100

def_df["Gole_katkı"] = df["Goals_s"]*70/100 + df["Assists_s"]*30/100
def_df["Dakika"] = df["Min_s"]
#new_df["new_rate"] = new_df["bloklar"]*30/100 + new_df["muda_muca"]*35/100 + new_df["pas"]* 25/100 + new_df["gole_katki"]*10/100
```

```python
def_df = def_df[(def_df["Pos"]=="DF")|(def_df["Pos"]== "MFDF")|(def_df["Pos"]== "DFFW")|(def_df["Pos"]== "DFMF")]
```

```python
def_df = info_df.merge(def_df, how = "left",on = "Id" )
```

```python
def_df.head()
```

```python
def_df.dropna(inplace=True)
```

```python
def_df = def_df[def_df["Mac"]>10]
```
