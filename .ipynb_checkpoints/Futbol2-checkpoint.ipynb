```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 500)
pd.set_option('display.float_format', lambda x: '%.4f' % x)
```

```python
df_ = pd.read_excel('/Users/huseyinefkanalp/Desktop/DataScience/DataSets/2022-2023Footbal.xls')
```

```python
df_.head()
```

```python
df = df_.copy()
df.fillna(0.0, inplace = True)
```

```python
## Defans
```

```python
df.reset_index(inplace=True)
```

```python
df.rename({"index": "Id"}, axis="columns", inplace = True)
```

```python
info_df = pd.DataFrame()
info_df["Id"] = df["Id"]
info_df["Name"] = df["Player"]
info_df["Nation"] = df["Nation"]
info_df["Team"] = df["Squad"]
info_df["League"] = df["Comp"]
info_df["Born"] = df["Born"]
info_df["Mac"] = df["MP"]
```

```python
df.drop(["Player","Nation","Squad","Comp","Born","MP"],inplace = True, axis = 1)
```

```python
for i in df.columns:
    if i == "Pos":
        continue
    else:
        df[i] = pd.to_numeric(df[i], errors='coerce')
```

```python
df.fillna(0.0, inplace = True)
```

```python
for i in df:
    if i == "Pos":
        continue
    else:
        df[i] = df[i].astype(float)
```

```python
for i in df:
    if (i == "Pos") | (i == "Id"):
        continue
    else:
        df[f"{i}_s"] = MinMaxScaler(feature_range = (0,1)).fit(df[[i]]).transform(df[[i]])
```

```python
df.head()
```

```python
def_df = pd.DataFrame()
def_df["Id"] = df["Id"]
def_df["Pos"] = df["Pos"]
#Bloklar
def_df["Blok"]= df["Blocks_s"]*40/100 + df["BlkSh_s"]*30/100 + df["BlkPass_s"]*30/100  
#Mdahale
def_df["Bask"]= df["Tkl_s"]*10/100 + df["TklWon_s"]*10/100 + df["TklDef3rd_s"]*14/100 +  \
df["TklDriAtt_s"]*16/100 + df["TklDri%_s"]*14/100 + \
df["TklDef3rd_s"]*16/100 + df["TklMid3rd_s"]*6/100 + df["TklW_s"]*14/100

def_df["Pas"] = df["PasTotPrgDist_s"]*8/100 + df["PasLonCmp_s"]*12/100 + df["PasAss_s"]*8/100 + df["Pas3rd_s"]*10/100 + \
df["PPA_s"]*12/100 + df["CrsPA_s"]*10/100 + df["PasAtt_s"]*10/100 + df["PasLive_s"]*10/100 + \
 df["Sw_s"]*10/100 + df["Rec_s"]*10/100

def_df["Gole_katk"] = df["Goals_s"]*70/100 + df["Assists_s"]*30/100
def_df["Dakika"] = df["Min_s"]
```

```python
def_df = def_df[(def_df["Pos"]=="DF")|(def_df["Pos"]== "MFDF")|(def_df["Pos"]== "DFFW")|(def_df["Pos"]== "DFMF")]
```

```python
def_df = info_df.merge(def_df, how = "left",on = "Id" )
```

```python
def_df = def_df.fillna(0.0)
```

```python
def_df.head()
```

```python
def_df.to_csv("defans_df2.csv")
```
